```python
# Import libaries
import pandas as pd
import requests
import folium
from geopy.geocoders import Nominatim

!pip install lxml
import lxml

!pip install bs4
import bs4 as bs
from bs4 import BeautifulSoup

print("Libraries Imported")
```

```python
# Recreate dataframe from Question 2
# Get data from the url and convert into dataframe
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
res = requests.get(url)
soup = bs.BeautifulSoup(res.content,features='html')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))
data = pd.read_json(df[0].to_json(orient='records'))

# Ignore cells with a borough that is not assigned
data = data[data.Borough != 'Not assigned']
data.head()

# Merge postal codes with multiple neighbourhoods
data = data.groupby(['Postal Code', 'Borough'])['Neighborhood'].apply(list).apply(lambda x:', '.join(x)).to_frame().reset_index()
data.head()

# If borough is not assigned, assign it the neighbourhood
for index, row in data.iterrows():
    if row['Neighborhood'] == 'Not assigned':
        row['Neighborhood'] = row['Borough']
        
# Create dataframe of coordinates
geospatial_url = "https://cocl.us/Geospatial_data"
geospatial_data = pd.read_csv(geospatial_url)

# Merge both dataframes
data = pd.merge(data, geospatial_data, on='Postal Code')

data
```

```python
# Get coordinates of Toronto on Geocoder
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="tr_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of the City of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# Create map of Toronto using latitude and longitude
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# Add markers to map
for lat, lng, borough, neighborhood in zip(data['Latitude'], data['Longitude'], data['Borough'], data['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
